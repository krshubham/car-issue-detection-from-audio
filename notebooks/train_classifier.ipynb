# Sound Issue Classifier Training

This notebook demonstrates how to train and use the sound classifier for identifying engine issues.

```python
import os
```

```python
import sys
sys.path.append('..')
from sound_classifier import SoundClassifier
```

```python
# Initialize and train the classifier
data_dir = '../data'
# Random Forest
rf_classifier = SoundClassifier(data_dir=data_dir, model_type='rf')
rf_classifier.train()

# Logistic Regression
lr_classifier = SoundClassifier(data_dir=data_dir, model_type='lr')
lr_classifier.train()

# Support Vector Machine
svm_classifier = SoundClassifier(data_dir=data_dir, model_type='svm')
svm_classifier.train()

# Neural Network (Multi-layer Perceptron)
nn_classifier = SoundClassifier(data_dir=data_dir, model_type='nn')
nn_classifier.train()
```

```python

# Save the trained model
lr_classifier.save_model('../models/lr_sound_classifier_model.joblib')
rf_classifier.save_model('../models/rf_sound_classifier_model.joblib')
svm_classifier.save_model('../models/svm_sound_classifier_model.joblib')
nn_classifier.save_model('../models/nn_sound_classifier_model.joblib')
classifiers = [lr_classifier, rf_classifier, svm_classifier, nn_classifier]
```

```python
test_file = '../test_data/air_filter_sample_6.wav'
for file in os.listdir('../test_data'):
    test_file = '../test_data/'+file
    print(test_file)
    for classifier in classifiers:
        print(f'Trained on {classifier.model_type} model')
        prediction = classifier.predict(test_file)
        print(f'Predicted issue: {prediction}')
    print('---'*5)
```
